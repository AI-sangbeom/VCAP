```python
import copy
def find_start_point(park):
    for i, h in enumerate(park):
        for j, w in enumerate(h):
            if park[i][j] == 'S':
                return [i, j]

def solution(park, routes):
    
    current_position = find_start_point(park)
    next_position = find_start_point(park)
    direction2num = [-1, 1]
    direction_w = ['W', 'E']
    direction_h = ['N', 'S']
    w_dict = dict(zip(direction_w, direction2num))
    h_dict = dict(zip(direction_h, direction2num))
    
    for route in routes:
        r = route.split(' ')
        direction, norm = r[0], int(r[1])
        if direction in direction_w:
            dw = w_dict[direction]
            dh = 0
        elif direction in direction_h:
            dw = 0
            dh = h_dict[direction]
        
        check = False
        print('current :', current_position)
        for i in [1]*norm:
            h = next_position[0] = next_position[0] + i*dh
            w = next_position[1] = next_position[1] + i*dw
            print('next :', next_position)
            # print(park[h][w])
            if h < 0 or w < 0 or park[h][w] == 'X':    
                check=True
                break
        
        if check == True:
            next_position = copy.deepcopy(current_position)
        else:
            current_position = copy.deepcopy(next_position)
            
        
    return current_position
```
